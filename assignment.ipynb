```python

wall_length = float(input('Enter the length of the wall (in meters): '))
wall_height = float(input('Enter the height of the wall (in meters): '))
cost_per_sq_meter = float(input('Enter the cost per square meter for painting (in ₹): '))
window_length = 1.5
window_height = 1.2
wall_area = wall_length * wall_height
one_window_area = window_length * window_height
total_window_area = 2 * one_window_area
area_to_be_painted = wall_area - total_window_area
total_cost = area_to_be_painted * cost_per_sq_meter
print(f'Total area to be painted: {area_to_be_painted:.2f} sq meters')
print(f'Total cost to paint the wall: ₹{total_cost:.2f}')
```

```python
import math
a = float(input('Enter side a: '))
b = float(input('Enter side b: '))
c = float(input('Enter side c: '))
s = (a + b + c) / 2
if s > a and s > b and s > c:
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))
    print(f'The area of the triangle is: {area:.2f}')
else:
    print('The given side lengths do not form a valid triangle.')
```
